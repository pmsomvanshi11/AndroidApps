```python
a = np.arange(4)
print(f'a: {a}')
c = a+2
print(f'c: {c}')
a[0] = 100
print(f'a: {a}')
print(f'c: {c}')
print(np.shares_memory(a,c)) # deep copy
```

```python
a = np.arange(4)
print(f'a: {a}')
b = a.reshape(2,2)
print(f'b: {b}')
a[0] = 100
print(f'a: {a}')
print(f'b: {b}')
print(np.shares_memory(a,b)) # shallow copy
```

```python
a = np.arange(10)
print(f'a: {a}')
b = a[::2]
print(f'b: {b}')
a[0] = 100
print(f'a: {a}')
print(f'b: {b}')
print(np.shares_memory(a,b))
```

```python
a = np.arange(6)
print(f'a: {a}')
b = a[a%1==0]# >> a%1==0 : bool array>> a[bool_array]=all those places where bool array is True
print(f'b: {b}')
a[0] = 100
print(f'a: {a}')
print(f'b: {b}')
print(np.shares_memory(a,b))
```

```python
# Memory usage in numpy:
#     Shallow copy(np.shares_memory==True): reshaping, slicing...
#     Deep copy (np.shares_memory==False): arithmetic operation, masking...
```

```python
a = np.arange(10)
print(f'a: {a}')
b = a.view() # shallow copy
print(f'b: {b}')
print(np.shares_memory(a,b))
```

```python
a = np.arange(10)
print(f'a: {a}')
b = a.copy() # deep copy
print(f'b: {b}')
print(np.shares_memory(a,b))
```

```python
a = np.array([0,1,2,3,4,5])
b = a[a%1 == 0]
b[0] = 10
a[:2]
```

```python
import copy
a = np.arange(10)
print(f'a: {a}')
b = copy.deepcopy(a) # deep copy
print(f'b: {b}')
print(np.shares_memory(a,b))
```

```python
a = np.arange(1,5)
b = np.arange(2,6)
c = np.arange(3,7)
print(f'a: \n{a}')
print('_'*25)
print(f'b: \n{b}')
print('_'*25)
print(f'c: \n{c}')
print('_'*25)
np.vstack([a,b,c])
```

```python
# concatenate
a = np.array([[1], [2], [3]])
b = np.array([[4], [5], [6]])
print(f'a: \n{a}')
print('_'*25)
print(f'b: \n{b}')
print('_'*25)
print(np.concatenate((a, b))) # by default, axis=0, vstack
print('_'*25)
print(np.concatenate((a, b), axis=1)) # axis=1, hstack
```

```python
# array >> shape
# a >> (_,_,_,*,_)
# b >> (_,_,_,*,_)
# * is axis=3

# axis=0>>(*,_)
# axis=1>>(_,*)
```

```python
a = np.array([[1], [2], [3]])
b = np.array([[4], [5], [6]])
print(f'a: \n{a}, ndim: {a.ndim}')
print('_'*25)
print(f'b: \n{b}, ndim: {b.ndim}')
```
